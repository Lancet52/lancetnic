# Classification. Training

```python
from lancetnic.models import LancetMCA
from lancetnic import Classification

model = Classification(text_column=['Sentiment'],
                       data_column=None,
                       label_column='category',
                       split_ratio=0.2,
                       random_state=42,
                       max_features=None)
model.train(model_name=LancetMCA,
            train_path="datasets/classification/COVID19/Corona_NLP_train_cut.csv",
            val_path=None,
            num_epochs=10,
            hidden_size=16,
            num_layers=1,
            batch_size=64,
            learning_rate=0.0001,
            dropout=0,
            optim_name='Adam',
            crit_name='CELoss',
            device_name='cpu'
            )           
            
```
